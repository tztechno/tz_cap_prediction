```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
img = cv2.imread('./data/IMG_9481.jpg')
plt.imshow(img)

v_split = 13  # 
h_split = 13  # 

_img = img[:h // v_split * v_split, :w // h_split * h_split]
out_imgs = []

for h_img in np.vsplit(_img, v_split):  # 
    for v_img in np.hsplit(h_img, h_split):  # 
        out_imgs.append(v_img)
out_imgs = np.array(out_imgs)
print(out_imgs.shape) 

fig, axes_list = plt.subplots(v_split, h_split, figsize=(5, 5))
axes_list = axes_list.ravel()
for i, (sub, axes) in enumerate(zip(out_imgs, axes_list)):
    axes.imshow(sub)
    axes.set_axis_off()
    axes.set_title(i)

plt.show()    
```

```python
image_data2 = []
    
for i in range(9481,9501):
    img = cv2.imread('./data/IMG_'+str(i)+'.jpg')
    plt.imshow(img)
    h, w, c = img.shape
    
    v_split = 13 # 
    h_split = 13 # 
    _img = img[:h // v_split * v_split, :w // h_split * h_split]

    image_data =[]

    for h_img in np.vsplit(_img, v_split):  # 
        for v_img in np.hsplit(h_img, h_split):  #     
            b, g, r = cv2.split(v_img)  # 
            image_data.append(round(g.mean())/100)

    image_data2.append(image_data)
                          
print(image_data2)

image_data3 = pd.DataFrame(image_data2)
image_data3.to_csv('image_data4.csv')
```
